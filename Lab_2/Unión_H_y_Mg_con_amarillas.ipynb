```python
archivos = ["Espectro_H.csv", "Esprectro_Mer.csv", "Linea_amarilla_del_Mer.csv"]
```

```python
file = '/content/drive/MyDrive/Laboratorio_moderna/Lab2/'
file
```

```python
plt.figure(figsize=(8,6))
for archivo in archivos:
    df = pd.read_csv(os.path.join(file, archivo))
    plt.plot(df.iloc[:,0], df.iloc[:,1], label=archivo)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Comparaci√≥n de los tres archivos")
plt.legend()
plt.grid(True)
plt.show()
```

```python
etiquetas = ["Hidr√≥geno", "Mercurio", "Mercurio + zoom"]
# Crear la figura solo una vez
fig, ax = plt.subplots(figsize=(10,6))
for archivo, etiqueta in zip(archivos, etiquetas):
    df = pd.read_csv(os.path.join(file, archivo))

    x = df.iloc[:,0]
    y = df.iloc[:,1]

    if "Linea_amarilla_del_Mer" in archivo:
        mask = x <= 1300
        x = x[mask]
        y = y[mask]
    if "Esprectro_Mer" in archivo:
        mask = x <= 1300
        x = x[mask]
        y = y[mask]

    ax.plot(x, y, label=etiqueta)

# Configuraci√≥n de la gr√°fica
ax.set_xlabel("Distancia (pixeles)")
ax.set_ylabel("Intensidad")
ax.set_title("Espectros H y Mg")
ax.legend(loc="lower left")
ax.grid(True)
# --- despu√©s de hacer la gr√°fica ---
plt.savefig("espectros_H_Mg.png", dpi=300, bbox_inches="tight")
plt.show()
```
