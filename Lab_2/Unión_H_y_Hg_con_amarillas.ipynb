```python
archivos = ["Espectro_H.csv", "Esprectro_Mer.csv", "Linea_amarilla_del_Mer.csv"]
```

```python
file = '/content/drive/MyDrive/Laboratorio_moderna/Lab2/'
file
```

```python
etiquetas = ["Hidrógeno", "Mercurio", "Mercurio + zoom"]
# Crear la figura solo una vez
fig, ax = plt.subplots(figsize=(10,6))
for archivo, etiqueta in zip(archivos, etiquetas):
    df = pd.read_csv(os.path.join(file, archivo))

    x = df.iloc[:,0]
    y = df.iloc[:,1]

    if "Linea_amarilla_del_Mer" in archivo:
        mask = x <= 1300
        x = x[mask]
        y = y[mask]
    if "Esprectro_Mer" in archivo:
        mask = x <= 1300
        x = x[mask]
        y = y[mask]

    ax.plot(x, y, label=etiqueta)

# Configuración de la gráfica
ax.set_xlabel("Distancia (pixeles)")
ax.set_ylabel("Intensidad")
ax.set_title("Espectros H y Hg")
ax.legend(loc="lower left")
ax.grid(True)
# --- después de hacer la gráfica ---
plt.savefig("espectros_H_Hg.png", dpi=300, bbox_inches="tight")
plt.show()
```
